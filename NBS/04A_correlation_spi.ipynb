### Figure out the top 4 variables which have the highest effect on spi_Score

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()
```

```python
correlation_matrix
```

```python
# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Get the absolute correlation values with 'spi_score'
correlation_with_spi = correlation_matrix['spi_score'].abs().sort_values(ascending=False)
```

```python
correlation_with_spi
```

```python
# Select the top 4 variables with the highest impact on 'spi_score'
top_4_impact_variables = correlation_with_spi[2:6]  # Excluding 'spi_score' and 'spi_rank'

top_4_impact_variables
```

```python
# Mention the variable names you want to include in the heatmap
top_4_variables = ['spi_score', 'access_info_comm', 'wellbeing', 'opportunity','access_adv_edu']
```

```python
# Filter the correlation matrix to include only the mentioned variables
filtered_corr_matrix = correlation_matrix.loc[top_4_variables, top_4_variables]
```

```python
# Create the heatmap for the mentioned variables using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(filtered_corr_matrix, annot=True, cmap='Blues', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap of Most Affecting Variables")
plt.show()
```
